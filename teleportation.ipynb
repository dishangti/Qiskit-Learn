```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_distribution
```

Transfer quantum state from $q_0$ to $q_2$

```python
qc = QuantumCircuit(3,3)
qc.x(0)     # q0 -> 1
qc.barrier()

# build an entanglement between q1 and q2
qc.h(1)
qc.cx(1,2)

# teleportation protocal
qc.cx(0,1)
qc.h(0)

# measurement
qc.barrier()
qc.measure([0,1], [0,1])

# recover
qc.barrier()
qc.cx(1,2)
qc.cz(0,2)

# measurement
qc.measure(2,2)

qc.draw(output='mpl')
```

```python
simulator = AerSimulator()
result = simulator.run(qc).result()
counts = result.get_counts()
plot_distribution(counts, title='State counts')
```
